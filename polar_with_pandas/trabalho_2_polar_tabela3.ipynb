```python
import pandas as pd

df = pd.read_csv("dados_tabela_3.csv")
df.head()
```

```python
df = df.rename(index=str, columns={' SO2':'SO2', ' NOx':'Nox',' CO':'CO',' HC':'HC',' MP10':'MP10'})
df.columns.values
```

```python
df.set_index(['Bacias Aéreas/Taxa de Emissão de Poluentes'])
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set('paper', 'white', 'colorblind', font_scale=2.2, 
            rc={'lines.linewidth': 2, 
                'figure.figsize': (12.0, 10.0), 
                'image.interpolation': 'nearest', 
                'image.cmap': 'gray'})

plt.rcParams['figure.figsize'] = (12.0, 10.0) 
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
```

```python
sns.barplot(x="Bacias Aéreas/Taxa de Emissão de Poluentes", y="SO2", data=df)

plt.ylabel("Taxa de Emissão (x1000 ton/ano)")
plt.xlabel("Nome da Bacia")

plt.xticks(rotation=45)

mean_poluente = df["SO2"].mean()
plt.axhline(mean_poluente,label='Média', linestyle='--')
legend = plt.legend(loc='upper right')

plt.savefig('tabela_3_SO2.png')
mean_poluente
```
